```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r'C:\Users\User\Desktop\Ironhack_DA\Project-2\df1_df2_joined_cleaned.csv'

df_final = pd.read_csv(file_path)
df_final = df_final.rename(columns={"AI Impact": "AI Impact %"})
```

```python
print(df_final.Job_Title.unique().tolist())
```

```python
# 1st insight:
ai_impact_by_risk = df_final.groupby("Automation_Risk")["AI Impact %"].mean()
ai_impact_by_risk
```

```python
# 1st insight graph
# Plot the data
ax = ai_impact_by_risk.plot(kind='bar', color=['r', 'g', 'b'])

# Adding a title and labels for clarity
plt.title('Average AI Impact by Automation Risk')
plt.xlabel('Automation Risk')
plt.ylabel('Average AI Impact')

# Add the numbers on top of the bars
for index, value in enumerate(ai_impact_by_risk):
    plt.text(index, value + 0.5, round(value, 2), ha='center')

# Show the plot
plt.show()

# conclusion: the average AI impact is the lowet in High automation risk
```

```python
# 3rd insight lets go explore the required skills because the 2nd insight needs further exploration
grouped_df_final_3 = df_final.groupby(["Job_Title","Required_Skills"]).agg({
    "AI_Adoption_Level": lambda x: x.mode()[0],
    "Automation_Risk": lambda x: x.mode()[0],
})

grouped_df_final_3_1 = grouped_df_final_3.loc[grouped_df_final_3.index.get_level_values('Job_Title').isin(["Software Engineer", "Data Scientist"])]
grouped_df_final_3_1

#consclusion about the skills: intersting part that Python as a skill has low risk for automation in data analysis compared to a software engineer role.
# Data analysis has low automation risk in both role, probably due to the nature of the job that requires story telling.
# Machine learnings is also intersting because in software engineers has high automation risk.

# Possible explanation: Data Scientists primarily use ML for analysis and decision-making,
# relying on human intuition and domain knowledge, leading to lower AI adoption and medium automation risk.
# In contrast, Software Engineers focus on building and automating ML applications, resulting in higher AI adoption and a greater risk of automation.
```

```python
# 3rd insight visulatisation to use in ppt

fig, ax = plt.subplots(figsize=(12, 4))  # Adjust width and height to fit content
ax.axis('off')  # Turn off the axis

# Create the table from the DataFrame
mpl_table = ax.table(
    cellText=grouped_df_final_2.values,
    colLabels=grouped_df_final_2.columns,
    rowLabels=grouped_df_final_2.index,
    cellLoc='center',
    loc='center'
)

# Styling
mpl_table.auto_set_font_size(False)
mpl_table.set_fontsize(10)
mpl_table.scale(1.2, 1.2)

# Header color
for (i, j), cell in mpl_table.get_celld().items():
    if i == 0 or j == -1:
        cell.set_text_props(weight='bold', color='white')
        cell.set_facecolor('#40466e')  # Header background color
    elif i % 2 == 0:
        cell.set_facecolor('#f0f0f0')  # Alternate row color

# Title
plt.title('Skills Characteristics Overview', fontsize=12, weight='bold', color='#40466e')

plt.show()
```

```python
# 4th insight: average workload ratio per Job title

grouped_df_final_4 = df_final.groupby("Job_Title").agg({
    "AI_Workload_Ratio": "mean"
})

# Convert AI_Workload_Ratio to percentage and round to two decimal places
grouped_df_final_4["AI_Workload_Ratio"] = grouped_df_final_4["AI_Workload_Ratio"] * 100
grouped_df_final_4["AI_Workload_Ratio"] = grouped_df_final_4["AI_Workload_Ratio"].round(2)

grouped_df_final_4 = grouped_df_final_4.sort_values(by="AI_Workload_Ratio", ascending=False)

# Display the updated DataFrame
grouped_df_final_4

# On average, only 13.41% of tasks in IT-related jobs are currently managed by AI. This suggests AI is augmenting rather than replacing the majority of IT work â€” aligning with your hypothesis that the fear may be overstated.
```

```python
sns.set_palette("viridis")

# Choose a color palette from seaborn
sns.set_palette("viridis")

# Plot the bar chart with the improved color scheme
plt.figure(figsize=(10, 6))
bars = plt.bar(grouped_df_final_4.index, grouped_df_final_4["AI_Workload_Ratio"], color=sns.color_palette("viridis", len(grouped_df_final_4)))

# Add labels on top of each bar with enhanced visibility
for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2, yval, f'{yval:.2f}%', 
        va='bottom', ha='center', fontsize=12, fontweight='bold', color='black'
    )

# Add titles and labels
plt.ylabel('AI Workload Ratio (%)')
plt.title('Average AI Workload Ratio by Job Title')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```
