```python
keywords = ['Data', 'Analyst', 'specialist', 'AI', 'UX', 'Product', 'ENGINEER', 'OPERATIONS']
keywords_lower = [re.escape(k.lower()) for k in keywords]

regex_pattern = '|'.join(keywords_lower)
#df1_filtered = df1[df1['Job_Title'].str.contains(regex_pattern, case=False, na=False)]
```

```python
df1['Job_Title'].apply(lambda x: re.findall(regex_pattern, x.lower())[0] if re.findall(regex_pattern, x.lower()) else None)
```

```python
df1['Job_Title_2'] = df1['Job_Title'].apply(lambda x: re.findall(regex_pattern, x.lower())[0] if re.findall(regex_pattern, x.lower()) else None)

df1
```

```python
replacements = {
    'Cyber Security Analyst': 'Cybersecurity Analyst',
    'Ux Designer': 'UX Designer',
    'Product Manager': 'Product Manager',
    'Software Engineer': 'Software Engineer',
    'Data Scientist': 'Data Scientist',
    'Operations Manager': 'Operations Manager'
}

# Assuming df1_cleaned is a view rather than a copy
df1_cleaned.loc[:, 'Job_Title'] = df1_cleaned['Job_Title'].replace(replacements)
```

```python
# Use .loc to avoid SettingWithCopyWarning
df1_cleaned.loc[:, 'AI Impact'] = (df1_cleaned['AI Impact'].astype(str))

df1_cleaned.loc[:, 'AI Impact'] = df1_cleaned['AI Impact'].str.strip().str.rstrip('%').replace('nan', None)

# Handle possible missing values before conversion
df1_cleaned.loc[:, 'AI Impact'] = df1_cleaned['AI Impact'].astype(int)
```
